```python
univ_ranking.loc[univ_ranking['rating'] != 0, :][:10].plot(
	kind='bar',
	x='name',
	y='userCount',
	color='purple',
	figsize=(20, 10)
)
plt.title('상위 10개 대학 별 유저수')
plt.show()
```

```python
univ_ranking.sort_values(by='solvedCount', ascending=False).loc[univ_ranking['rating'] != 0, :][:10].plot(
	kind='bar',
	x='name',
	y='solvedCount',
	color='purple',
	figsize=(20, 10)
)
plt.title('문제 해결 수 상위 10개 대학 별 문제 해결 수')
plt.show()
```

## 3. 단국대학교 랭킹 정보 데이터 분석하기

```python
dankook_univ = pd.DataFrame(solved_ac.getUnivUserRanking('단국대학교'))
dankook_univ = dankook_univ[
	['nickname', 'bio', 'solvedCount', 'voteCount', '_class', 'classDecoration', 'tier', 'rating',
	 'ratingByProblemsSum',
	 'ratingByClass', 'ratingBySolvedCount', 'ratingByVoteCount', 'exp', 'rivalCount',
	 'reverseRivalCount', 'maxStreak',
	 'rank', 'globalRank']]
dankook_univ
```

```python
dankook_univ_tier = dankook_univ
dankook_univ_tier['count'] = 1

dankook_univ_tier = dankook_univ_tier.groupby('tier').sum()
dankook_univ_tier

dankook_univ_tier['count'].plot(
	kind='pie',
	autopct='%1.1f%%',
	startangle=0,
	colors=['red', 'blue', 'green', 'yellow', 'silver', 'brown', 'gray'],
	figsize=(20, 10)
)
plt.title('단국대학교 티어 비율')
plt.show()
```

```python
dankook_univ_class = dankook_univ
dankook_univ_class['count'] = 1

dankook_univ_class = dankook_univ_class.groupby('_class').sum()
dankook_univ_class

dankook_univ_class['count'].plot(
	kind='pie',
	autopct='%1.1f%%',
	startangle=0,
	colors=['red', 'blue', 'green', 'yellow', 'silver', 'brown', 'gray'],
	figsize=(20, 10)
)
plt.title('단국대학교 클래스 비율')
plt.show()
```

```python
dankook_univ.plot(
	kind='scatter',
	x='solvedCount',
	y='rating',
	marker='+',
	cmap='viridis',
	c='yellow',
	s=100,
	alpha=0.3,
	figsize=(20, 10)
)
plt.title('단국대학교 문제 해결 수 별 점수')
plt.show()
```

```python
dankook_univ.groupby(['tier', '_class']).sum().plot(
	kind='barh',
	y='count',
	width=0.5,
	color=['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'black'],
	figsize=(20, 10)
)
plt.title('단국대학교 티어와 클래스')
plt.show()
```
